```python
african_recipe_urls = []
for recipe in african_recipes_list:
    url = "https://www.bbc.com" + recipe.get('href')
    african_recipe_urls.append(url)

african_recipe_urls
```

```python
african_ingredients = []
for URL in african_recipe_urls:
    recipe = requests.get(URL)
    recipe_text = recipe.text
    recipe_soup = BeautifulSoup(recipe_text)
    ingredients_list = recipe_soup.findAll('li', {'class':"recipe-ingredients__list-item"})
    ingredients = [ingredient.text for ingredient in ingredients_list]
    african_ingredients.append(ingredients)
african_ingredients
```

```python
african_recipe_names = []
for URL in african_recipe_urls:
    recipe = requests.get(URL)
    recipe_text = recipe.text
    recipe_soup = BeautifulSoup(recipe_text)
    recipe_name = recipe_soup.find('h1', {'class':'gel-trafalgar content-title__text'})
    african_recipe_names.append(recipe_name.text)
```

```python
african_ingredients
```

```python
recipe_and_ingredients = list(zip(african_recipe_names, african_ingredients))
recipe_and_ingredients
```

```python
recipe_and_ingredient_paragraph = []
for recipe in recipe_and_ingredients:
    concat_recipe = ' '.join(recipe[1])
    recipe_name_ingredient = (recipe[0], concat_recipe)
    recipe_and_ingredient_paragraph.append(recipe_name_ingredient)

recipe_and_ingredient_paragraph
```

```python
import pandas as pd
africa_recipe= pd.DataFrame(recipe_and_ingredient_paragraph[1:])
newcol= 'African'
africa_recipe= africa_recipe.assign(target=newcol)
africa_recipe= africa_recipe.rename(columns={0: 'Recipe_Name', 1: 'Recipe_Ingredients'})
africa_recipe
```
