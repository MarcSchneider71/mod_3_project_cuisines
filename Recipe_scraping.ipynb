```python
african_recipe_urls = []
for recipe in african_recipes_list:
    url = "https://www.bbc.com" + recipe.get('href')
    african_recipe_urls.append(url)

african_recipe_urls
```

```python
african_ingredients = []
for URL in african_recipe_urls:
    recipe = requests.get(URL)
    recipe_text = recipe.text
    recipe_soup = BeautifulSoup(recipe_text)
    ingredients_list = recipe_soup.findAll('li', {'class':"recipe-ingredients__list-item"})
    ingredients = [ingredient.text for ingredient in ingredients_list]
    african_ingredients.append(ingredients)
african_ingredients
```

```python
african_recipe_names = []
for URL in african_recipe_urls:
    recipe = requests.get(URL)
    recipe_text = recipe.text
    recipe_soup = BeautifulSoup(recipe_text)
    recipe_name = recipe_soup.find('h1', {'class':'gel-trafalgar content-title__text'})
    african_recipe_names.append(recipe_name.text)
```

```python
cuisines = ['african', 'american', 'british', 'caribbean', 'chinese', 'east_european', 'french', 
            'greek', 'indian', 'irish', 'italian', 'japanese', 'korean', 'mexican', 'nordic', 
            'north african', 'pakistani', 'portuguese', 'south_american', 'spanish',
            'thai_and_south-east_asian', 'turkish_and_middle_eastern']
```

```python
cuisines_recipe_urls = []
name_ingredients_cuisine = []
for cuisine in cuisines:
    cuisine_recipes = requests.get(f'https://www.bbc.com/food/cuisines/{cuisine}')
    cuisine_recipes_text = cuisine_recipes.text
    cuisine_recipes_soup = BeautifulSoup(cuisine_recipes_text)
    cuisines_recipes_list = cuisine_recipes_soup.findAll('div', {'class':"gel-layout__item gel-1/2 gel-1/3@m gel-1/4@xl"})
    
    for recipe in cuisines_recipes_list:
        url = "https://www.bbc.com" + recipe.findNext('a').get('href')
        #gives us a list of all recipe URLs
        cuisines_recipe_urls.append(url)
        
    for URL in cuisines_recipe_urls:
        recipe = requests.get(URL)
        recipe_text = recipe.text
        recipe_soup = BeautifulSoup(recipe_text)
        recipe_name = recipe_soup.find('h1', {'class':'gel-trafalgar content-title__text'}).text

        ingredients_list = recipe_soup.findAll('li', {'class':"recipe-ingredients__list-item"})
        ingredients = ' '.join([ingredient.text for ingredient in ingredients_list])
        
        n_i_c = (recipe_name, ingredients, cuisine)
        name_ingredients_cuisine.append(n_i_c)
        

        
        
# print(name_ingredients_cuisine)[1]
        
#     df = pd.DataFrame(name_ingredients_cuisine)[1:]

# df
# import pandas as pd
# africa_recipe= pd.DataFrame(recipe_and_ingredient_paragraph[1:])
# newcol= 'African'
# africa_recipe= africa_recipe.assign(target=newcol)
# africa_recipe= africa_recipe.rename(columns={0: 'Recipe_Name', 1: 'Recipe_Ingredients'})
# africa_recipe
```
